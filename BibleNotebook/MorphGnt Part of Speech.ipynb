# MorphGNT Part of Speech

Extract Part of Speech from MorphGNT

## LoadParse MorphGnt File into DataFrame (DF_WORDS)

Loads the `morphgnt.csv` into `DF_WORDS`

```python
import pandas as pd
from pprint import pprint

DF_WORDS = pd.read_csv("morphgnt.csv", index_col="Index")

print("===== DF_WORDS")
print(DF_WORDS.__class__.__name__)
print("-----")
pprint(vars(DF_WORDS))
print("-----")
pprint(DF_WORDS)
```

## Get Part of Speech for each Lemma

```python
GB_LEMMA = DF_WORDS.groupby(["Lemma"])

# print("===== GB_LEMMA")
# print(GB_LEMMA.__class__.__name__)
# print("-----")
# pprint(vars(GB_LEMMA))
# print("-----")
# pprint(GB_LEMMA)

list = [(name[0], group["Part of Speech"].unique()[0]) for name, group in GB_LEMMA]
dict(list)
# for name, group in GB_LEMMA:
#     print(name[0])
#     print(group["Part of Speech"].unique()[0])
#     print('----')
```

## Extract Lemma Series from DataFrame (S_LEMMA)

```python
S_LEMMA = DF_WORDS["Lemma"]

print("===== S_LEMMA")
print(S_LEMMA.__class__.__name__)
print("-----")
pprint(vars(S_LEMMA))
print("-----")
pprint(S_LEMMA)
```

## Group Lemmas by Value (GB_LEMMA_VALUES)

```python
GB_LEMMA_VALUES = S_LEMMA.groupby(S_LEMMA.values)

print("===== GB_LEMMA_VALUES")
print(GB_LEMMA_VALUES.__class__.__name__)
print("-----")
pprint(vars(GB_LEMMA_VALUES))
print("-----")
pprint(GB_LEMMA_VALUES)
```

## Get Word Count for Each Lemma (S_LEMMA_VALUE_COUNTS)

```python
S_LEMMA_VALUE_COUNTS = GB_LEMMA_VALUES.count()

print("===== S_LEMMA_VALUE_COUNTS")
print(S_LEMMA_VALUE_COUNTS.__class__.__name__)
print("-----")
pprint(vars(S_LEMMA_VALUE_COUNTS))
print("-----")
pprint(S_LEMMA_VALUE_COUNTS)
```

```python
DF_ANALYSIS = S_LEMMA_VALUE_COUNTS.to_frame(name="Lemma Count")
DF_ANALYSIS.index.name = "Lemma"

print("===== DF_ANALYSIS")
print(DF_ANALYSIS.__class__.__name__)
print("-----")
pprint(vars(DF_ANALYSIS))
print("-----")
pprint(DF_ANALYSIS)
```
